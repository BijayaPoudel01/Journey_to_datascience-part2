```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Line plot = line plot are effective for showing relationships,patterns, and trends between two numerical variables.

```python
day=['sunday','monday','tuesday','wednesday','thursday','friday','saturday']
temp=[35,26,40,23,34,21,37]
plt.plot(day,temp)
```

## Plt.scatter() =gives in point

```python
plt.scatter(day,temp)
```

## title,xlabel,ylabel

```python
plt.scatter(day,temp)
plt.xlabel('days')
plt.ylabel('temperature')
plt.title("week temperature")
```

# marker

```python
plt.figure(figsize=(10,4))
plt.xticks(df['index'])
plt.plot(df['index'],df['RG Sharma'],marker='>',markersize=10,color='green',linewidth=4,linestyle='dashed')
plt.title('rohit sharma')
plt.xlabel('season')
plt.ylabel('runs')
```

## plot more than 2 lines in a single charts

```python
plt.figure(figsize=(10,4))
plt.xticks(df['index'])
plt.plot(df['index'],df['RG Sharma'],marker='>',markersize=10,color='green',linewidth=4,linestyle='dashed')
plt.plot(df['index'],df['V Kohli'],marker='*',markersize=10,color='blue',linewidth=3,linestyle='solid')
plt.title('rohit sharma')
plt.xlabel('season')
plt.ylabel('runs')
```

## Legend(IMP)

```python
plt.figure(figsize=(10,4))
plt.xticks(df['index'])
plt.plot(df['index'],df['RG Sharma'],marker='>',markersize=10,color='green',linewidth=4,linestyle='dashed',label='rohit')
plt.plot(df['index'],df['V Kohli'],marker='*',markersize=10,color='blue',linewidth=3,linestyle='solid',label='kohli')
plt.legend(loc='upper left')
plt.title('rohit sharma')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
plt.figure(figsize=(10,4))
plt.xticks(df['index'])
plt.plot(df['index'],df['RG Sharma'],marker='>',markersize=10,color='green',linewidth=4,linestyle='dashed',label='rohit')
plt.plot(df['index'],df['V Kohli'],marker='*',markersize=10,color='blue',linewidth=3,linestyle='solid',label='kohli')
plt.legend(loc='lower center')
plt.title('rohit sharma')
plt.xlabel('season')
plt.ylabel('runs')
plt.grid()
plt.show()
```

## axhline and axvline

```python
plt.figure(figsize=(10,4))
plt.xticks(df['index'])
plt.plot(df['index'],df['RG Sharma'],marker='>',markersize=10,color='green',linewidth=4,linestyle='dashed',label='rohit')
plt.plot(df['index'],df['V Kohli'],marker='*',markersize=10,color='blue',linewidth=3,linestyle='solid',label='kohli')
plt.legend(loc='lower center')
plt.title('rohit sharma')
plt.xlabel('season')
plt.ylabel('runs')
plt.axvline(2010,color='yellow') #vertical line
plt.axhline(850,color='yellow') #horizontal line
plt.grid()
plt.show()
```
