## SQL Joins

```python
import pandas as pd
```

```python
df_employees=pd.DataFrame({
'emp_id':[1,2,3,4],
'name':['Biju','Ram','Hari','Shital'],
'dep_id':[101,102,103,104]
})
df_employees
```

```python
df_departments=pd.DataFrame({
    'dep_id':[101,102,105],
    'dep_name':['HR','Finance','IT']
})
df_departments
```

## Inner Join

```python
pd.merge(df_employees,df_departments,on='dep_id',how='inner') #gives details of biju and ram.i,e. common dep_id.
```

### Left Join

```python
pd.merge(df_employees,df_departments,on='dep_id',how='left') #gives details of 1st tebale all along with dep name of biju and ram as well.
```

## Right Join

```python
pd.merge(df_employees,df_departments,on='dep_id',how='right') #gives details of 2nd table all along with name of biju and ram as well.
```

## Full outer join

```python
pd.merge(df_employees,df_departments,on='dep_id',how='outer') #gives details of all like union
```

### Different columns name

```python
pd.merge(df_employees, df_departments, how='inner', left_on='dep_id', right_on='dep_id')
```

# Concatenate

```python
ids=[1,2,3,4,5]
product=['headphone','mobile','laptop','ipad','airpod']
unit_price=[10000,20000,30000,40000,50000]
quantity=[10,12,14,5,3]
dicts={'id':ids,'Product':product,'UnitPrice':unit_price,'Quantity':quantity}
dicts
```

```python
jan=pd.DataFrame(dicts)
jan
```
